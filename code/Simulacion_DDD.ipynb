```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def calcular_tiempo(x, x1, y1, x2, y2, v1, v2):
    t1 = np.sqrt((x1 - x)**2 + y1**2) / v1
    t2 = np.sqrt((x2 - x)**2 + y2**2) / v2
    return t1 + t2
```

```python
# Coordenadas de los puntos P1 y P2
x1, y1 = 2, 2
x2, y2 = 10, -2

# Índices de refracción de los medios
n1 = 1.50 #Acrílico
n2 = 2.42 #Diamante

# Velocidades de la luz en cada medio
v1 = 3e8 / n1
v2 = 3e8 / n2
```

```python
# Rango de valores para la posición x
x_valores = np.linspace(0, 10, 100)

# Cálculo del tiempo para cada valor de x
valores_tiempo = [calcular_tiempo(x, x1, y1, x2, y2, v1, v2) for x in x_valores]
```

```python
# Obtener el índice del valor mínimo
indice_minimo = np.argmin(valores_tiempo)
x_minimo = x_valores[indice_minimo]
tiempo_minimo = valores_tiempo[indice_minimo]
```

```python
def format_y_axis(value, pos):
    return '{:.2e}'.format(value)
```

```python
plt.ticklabel_format(style='plain', axis='both', useOffset=False)
plt.gca().xaxis.set_major_formatter(plt.FormatStrFormatter('%.2f'))

# Gráfica del tiempo versus posición x
plt.plot(x_valores, valores_tiempo)
plt.scatter(x_minimo, tiempo_minimo, color='red', label='Posición en el mínimo tiempo')

# Etiqueta del punto mínimo
plt.text(x_minimo, tiempo_minimo, f'({x_minimo:.2f}, {tiempo_minimo:.2f})',
         color='red', ha='center', va='bottom')

# Detalles de la gráfica
plt.xlabel('Posición (m)')
plt.ylabel('Tiempo (s)')
plt.title('Gráfica Tiempo vs Posición')
plt.grid(True)
plt.legend()

# Formato del eje Y
plt.ticklabel_format(axis='y', style='plain')
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(format_y_axis))

# Mostrar el valor mínimo de tiempo
print(f"El tiempo mínimo es {tiempo_minimo:.6e} segundos en la posición x = {x_minimo:.2f} m")

plt.show()
```
